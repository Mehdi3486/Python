```python
 def linksoup(link) : 
    import requests
    from bs4 import BeautifulSoup
    dicti={}
    lis=[]
    li=[]
    lih2=[]
    lih3=[]
    lih4=[]
    paraf=[]
    key=[]
    lislink=[]
#    link = "https://fr.wikipedia.org/wiki/Adjectif"
    page = requests.get(link)
    content = page.content
    soup = BeautifulSoup(content,"html.parser")
    container = soup.find("div",{"id":"mw-content-text"})
    paras = container.find_all("p")
    head1 = container.find_all("h1")
    head2 = container.find_all("h2")
    head3 = container.find_all("h3")
    head4 = container.find_all("h4")
    linka = container.find_all("a")
    texte = soup.title.string
    con = soup.find("h1",{"class":"firstHeading mw-first-heading"}).text
    t=len(paras)
    lh2=len(head2)
    lh3=len(head3)
    lh4=len(head4)
    lin1=len(linka)
    for linki in range (lin1):
        lislink.append(linka[linki].attrs)
        #print(linka[linki].attrs)
    #for h2 in range (lh2):
    #    print(head2[h2].text)
    #for h3 in range (lh3):
    #    print(head3[h3].text)
    #for h4 in range (lh4):
    #    print(head4[h4].text)
    for q in range (t):
        li.append(paras[q].text)
    for ch2 in range(lh2):
        lih2.append(head2[ch2].text)
    for ch3 in range(lh3):
        lih3.append(head3[ch3].text)
    for ch4 in range(lh4):
        lih4.append(head4[ch4].text)
    ner = soup.find("div",{"class":"mw-parser-output"})
    e=0
    s=0
    for tag in ner.find_all(True): 
        if tag.name=='p':
            lis.append(tag.name)
        elif tag.name[0]=='h':
            lis.append(tag.name)
        else:
            pass
    lis.append('av')
    f=len(lis)
    pp=0
    ph2=0
    ph3=0
    ph4=0
    pf=0
    hed=0
    stock=''
    st=0
    key.append(texte)
    for j in range (f-1):
        if lis[j]=='p' and lis[j+1]=='p':
            stock=stock + li[pp]
            pp=pp+1
            st=st+1           
        elif lis[j]=='p' and lis[j+1]!='p':
            if stock!='':
                stock=stock + li[pp]
                paraf.append(stock)
                stock=''
                st=0
            else:
                paraf.append(li[pp])
            pp=pp+1
        elif lis[j]=='h2' and lis[j+1]!='p':
            ph2=ph2+1
        elif lis[j]=='h2' and lis[j+1]=='p':
            key.append(lih2[ph2])
            ph2=ph2+1
        elif lis[j]=='h3' and lis[j+1]!='p':
            ph3=ph3+1
        elif lis[j]=='h3' and lis[j+1]=='p':
            key.append(lih3[ph3])
            ph3=ph3+1
        elif lis[j]=='h4' and lis[j+1]!='p':
            ph4=ph4+1
        elif lis[j]=='h4' and lis[j+1]=='p':
            key.append(lih4[ph4])
            ph4=ph4+1
        else:
            pass
    lp=len(paraf)
    lk=len(key)
    if lp==lk:
        lm=lk
    elif lp<lk:
        lm=lp
    else:
        lm=lk
    for lop in range(lm):
        dicti[key[lop]] = paraf[lop]
    return dicti ,lislink
#linksoup("https://fr.wikipedia.org/wiki/Adjectif")
linksoup("https://fr.wikipedia.org/wiki/Python_(langage)")
```
